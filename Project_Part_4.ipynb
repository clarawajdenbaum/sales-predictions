```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filename = '/content/drive/MyDrive/Project/sales_predictions.csv'
df = pd.read_csv(filename)
corr = df.corr()
```

```python
# important codes from Project - Part 2
weight_mean = df['Item_Weight'].mean()
df['Item_Weight'].fillna(weight_mean, inplace=True)
df['Outlet_Size'].fillna('Missing', inplace=True) 
df['Item_Fat_Content'] = df['Item_Fat_Content'].replace('LF','Low Fat')
df['Item_Fat_Content'] = df['Item_Fat_Content'].replace('low fat', 'Low Fat')
df['Item_Fat_Content'] = df['Item_Fat_Content'].replace('reg', 'Regular')
```

```python
df.head(2)
```

```python
sns.heatmap(corr, cmap = 'Pastel1_r', annot = True);

# With this heatmap, we can see that there is a correlation of 0.57 between Item_MRP and Item_Outlet_Sales.
# 0.57 is a moderate correlation.
```

```python
# Properties and outlets of products

# The goal is to understand the properties of products and outlets that play crucial roles in increasing sales.
```

```python
plt.style.use('seaborn');
df.boxplot(column = 'Item_Outlet_Sales', by = 'Item_Type', color='magenta', medianprops=dict(color="blue", alpha=0.9));
plt.xticks(rotation=90);
plt.ylabel('Item Outlet Sales', fontsize = 12);
plt.xlabel('Item Type', fontsize = 12);
plt.title('Item Outlet Sales by item Type');
plt.suptitle('');
plt.grid(axis='y', c='w', alpha=0.7, linestyle='-')
# The boxplot that I have created shows the item outlet sales by the item type.
# I want to be able to visualize the difference in sales of each item type. 
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(8,10));
df['Item_Outlet_Sales'].hist(alpha=.5, color = 'magenta', edgecolor = 'Black', bins= 'auto')
plt.xlabel('Item Outlet Sales', fontsize=15) 
plt.ylabel('Frequency', fontsize=15);
plt.title('Frequency of Item Outlet Sales', fontsize=15);

# This histogram shows the frequency of the item outlet sales. 
# I want to have a clear visualisation of the sales' amount frequency.
```

```python
df.hist(column='Item_Outlet_Sales', by='Item_Fat_Content', bins='auto', grid=False, color='magenta', alpha=.5, edgecolor = 'Black');
plt.xlabel('Item Outlet Sales', fontsize=15) 
plt.ylabel('Frequency',fontsize=15);

# Here we can see the variation of the frequency of the item outlet sales by the item fat content (low fat or regular).
```
